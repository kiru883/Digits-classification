```
# data augmentation(rotate on 25 grad.)
def get_bounding_rect(obj):
    x_min, x_max = obj.shape[1], 0
    y_min, y_max = obj.shape[0], 0
    for y in range(obj.shape[0]):
        for x in range(obj[y].shape[0]):
            if obj[y][x] != 0:
                if x_min > x:   x_min = x
                if x_max < x:   x_max = x
                if y_min > y:   y_min = y
                if y_max < y:   y_max = y

    bounded_image = obj[y_min:y_max+1, x_min:x_max+1]
    return bounded_image

def get_mnist_image(obj, angle, noise_coef=5):
    # get bound
    image = get_bounding_rect(obj.reshape(28, 28))
    # rotate
    image = sp.ndimage.rotate(image, angle)
    # noise deliverance
    image -= noise_coef
    image = np.where(image < 0, 0, image) 
    # get bound
    image = get_bounding_rect(image)
    # resize on max size
    if image.shape[0] >= image.shape[1]:
        x_side = 20
        y_side =  int(np.around((20 * image.shape[1]) / image.shape[0]))
    else:
        x_side = int(np.around((20 * image.shape[0]) / image.shape[1]))
        y_side = 20
    
    image = resize(image, (x_side, y_side))
    image = np.around(MinMaxScaler().fit_transform(image) * 255).astype(int)
    mnist = np.zeros((28, 28))
    half_y = y_side//2
    half_x = x_side//2
    mnist[14-(half_x):14+(x_side-half_x), 14-(half_y):14+(y_side-half_y)] = image
    return mnist


# get rotated digits
noise_coef = 32
test_objects_num = 30000
angle = 24

dataset = pd.concat([digits_train, digits_test], axis=0)
x_objects = []
y_objects = []
for obj_x, obj_y in zip(dataset.drop('label', axis=1).to_numpy(), dataset['label']):
    x_objects.append(get_mnist_image(obj_x, 0, noise_coef).reshape(-1,))
    x_objects.append(get_mnist_image(obj_x, -angle, noise_coef).reshape(-1,))
    x_objects.append(get_mnist_image(obj_x, angle, noise_coef).reshape(-1,))
    y_objects += 3*[obj_y]

x_objects = np.array(x_objects)
y_objects = np.array(y_objects)
ind = np.random.permutation(x_objects.shape[0])
x_objects = x_objects[ind]
y_objects = y_objects[ind]

x_train, y_train = x_objects[test_objects_num:], y_objects[test_objects_num:]
x_test, y_test = x_objects[:test_objects_num], y_objects[:test_objects_num]
x_train = MinMaxScaler().fit_transform(x_train)
x_test = MinMaxScaler().fit_transform(x_test)

del x_objects, y_objects, ind, dataset, test_objects_num
gc.collect()
```

```
# select best gb model of xgb, lgb and catboost 
n_estimators = 100

train_obj_num = 60000
accs_arr = []

#catboost
cgb_model = cgb.CatBoostClassifier(n_estimators=n_estimators, verbose=0)
cgb_model.fit(x_train[:train_obj_num], y_train[:train_obj_num])
acc = accuracy_score(np.argmax(cgb_model.predict_proba(x_test), axis=1), y_test)
accs_arr.append(acc)

#xgboost
xgb_model = xgb.XGBClassifier(n_estimators=n_estimators)
xgb_model.fit(x_train[:train_obj_num], y_train[:train_obj_num])
acc = accuracy_score(np.argmax(xgb_model.predict_proba(x_test), axis=1), y_test)
accs_arr.append(acc)

#lightGbm
lgb_model = lgb.LGBMClassifier(n_estimators=n_estimators)
lgb_model.fit(x_train[:train_obj_num], y_train[:train_obj_num])
acc = accuracy_score(np.argmax(lgb_model.predict_proba(x_test), axis=1), y_test)
accs_arr.append(acc)

_, ax = plt.subplots(figsize=(18, 12))
sns.barplot(x=['catboost', 'xgboost', 'lightgbm'], y=accs_arr, ax=ax)
ax.set_title('accuracy / model (100 trees)', fontsize=20)
ax.set_xlabel('model', fontsize=20)
ax.set_ylabel('accuracy', fontsize=20)
```

```
# find best activ./loss functions, softmax layer, bias hyperparams

input_vec = 784
classes = 10

#hyperparam space for NN
space_stage1 = {
    "layers": [
               input_vec,
               30,
               classes
    ],
    "alpha":  0.3,
    "epochs": 4,
    "activate_type":  hp.choice('activate_type', [
                                                  'logistic',
                                                  'relu',
                                                  'tanh',
                                                  'softsign',
                                                  'elu',
                                                  'softplus',
                                                  'lrelu',
                                                  'swish'
    ]),
    "loss_type":  hp.choice('loss_type', [
                                          'multiclassEntropy',
                                          'mse',
                                          'helinger',
                                          'kullback'
    ]),
    "batch_size": 30,
    "softmax_output": hp.choice('softmax_output', [True, False]),
    "bias": hp.choice('bias', [True, False]),
    "verbosity":  False
}

# init. functions for minimaze
def objective(params):
    clf = FNN(**params)
    clf.fit(x_train, y_train)

    predicts = clf.predict_proba(x_test)
    score = 0
    for ind in range(len(predicts)):
        if y_test[ind] == predicts[ind].argmax():
            score += 1
    acc = score/y_test.shape[0]
    print("Parameters: ", params, " Accuracy: ", acc)
    return -acc

#get best loss/activ/softmax and bias hyperparams.
best = fmin(fn=objective,
            space=space_stage1,
            algo=tpe.suggest,
            max_evals=130)
best_params = space_eval(space_stage1, best)
print(best_params)
```

```
# CNN activations after pooling layers
def reset_weights(model):
    session = get_session()
    for layer in model.layers: 
        if hasattr(layer, 'kernel_initializer'):
            layer.kernel.initializer.run(session=session)

activ = [[True, False], [False, True], [True, True], [False, False]]
accs_arr = []
f = 30

train_obj_num = 50000

X = np.concatenate([x_test, x_train])
y = np.concatenate([y_test, y_train])

for a in activ:
    model = Sequential()
    model.add(Conv2D(filters=f, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    # activ. layer 1
    if a[0]:
        model.add(Activation(activation='relu'))

    model.add(Conv2D(filters=2*f, kernel_size=(3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    # activ. layer 2
    if a[1]:
        model.add(Activation(activation='relu'))

    model.add(Flatten())
    model.add(Dense(units=30, activation='relu'))
    model.add(Dense(units=15, activation='relu'))
    model.add(Dense(units=10, activation='softmax')) 

    model.compile(
        optimizer='sgd',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    a = 0
    for i in range(10): 
        ind = np.random.permutation(X.shape[0])
        X_shuffled, y_shuffled = X[ind], y[ind]
        xtest, xtrain = X_shuffled[:train_obj_num], X_shuffled[train_obj_num:]
        ytest, ytrain = y_shuffled[:train_obj_num], y_shuffled[train_obj_num:]

        model.fit(xtrain.reshape(-1, 28, 28, 1), to_categorical(ytrain, num_classes=10),
                epochs=2,
                batch_size=32,
                verbose=0)
        a += model.evaluate(xtest.reshape(-1, 28, 28, 1), to_categorical(ytest, num_classes=10), verbose=0)[1]
        reset_weights(model)
        del X_shuffled, y_shuffled, xtest, xtrain, ytest, ytrain

    acc = a / 10
    accs_arr.append(acc)
    

_, ax = plt.subplots(figsize=(15, 10))
sns.barplot(x=['A1-none', 'none-A2', 'A1-A2', 'none-none'], y=accs_arr, ax=ax)
ax.set_xlabel('activations availability', fontsize=20)
ax.set_ylabel('accuracy', fontsize=20)
ax.set_title('Accuracy / activations availability on layers', fontsize=20)
ax.set_ylim([min(accs_arr)-0.02, 1])
```

```
# get accuracy train/test for each model (6 test folds), 1 valid fold
n_folds = 6
models = {
    "DNN": dnn_model,
    "CNN": cnn_model,
    "GB": gb_model
}

df = pd.DataFrame()
fold_size = round(x_train.shape[0] / n_folds)
train_indx = [(ind, ind+fold_size) for ind in range(0, x_train.shape[0], fold_size)]

for model in models.keys():
    acc_train, acc_test = 0, 0
    for fold_i, fold_e in train_indx:
        if model == 'CNN':
            xtrain, xtest = x_train.reshape(-1, 28, 28, 1), x_test.reshape(-1, 28, 28, 1)
            ytrain, ytest = np.copy(y_train), np.copy(y_test)
        else:
            xtrain, xtest = np.copy(x_train), np.copy(x_test)
            ytrain, ytest = np.copy(y_train), np.copy(y_test)

        acc_train += accuracy_score(ytrain[fold_i:fold_e],
                                    np.argmax(models[model].predict_proba(xtrain[fold_i:fold_e]), axis=1))
        
    acc_train /= n_folds    
    acc_test = accuracy_score(ytest, np.argmax(models[model].predict_proba(xtest), axis=1))
    
    model_df = pd.DataFrame({
        "accuracy": [acc_train, acc_test],
        "model": model,
        "type": ["train", "test"]
    })
    df = pd.concat([df, model_df], axis=0)
    del xtrain, xtest, ytrain, ytest, model_df

_, ax = plt.subplots(figsize=(18, 12))
sns.barplot(x="model", y="accuracy", hue="type", data=df, ax=ax)
ax.set_title("Models accuracy", fontsize=20)
ax.set_ylim([df['accuracy'].min()-0.015, 1])
```

```
# select ensambles
df = pd.DataFrame()
#get train predicts
dnn_predicts_train = dnn_model.predict_proba(x_train)
cnn_predicts_train = cnn_model.predict_proba(x_train.reshape(-1, 28, 28, 1))
gb_predicts_train = gb_model.predict_proba(x_train)
#get test predicts
dnn_predicts_test = dnn_model.predict_proba(x_test)
cnn_predicts_test = cnn_model.predict_proba(x_test.reshape(-1, 28, 28, 1))
gb_predicts_test = gb_model.predict_proba(x_test)
#predict matrixs
predicts_train = np.concatenate([dnn_predicts_train, cnn_predicts_train, gb_predicts_train], axis=1)
predicts_test = np.concatenate([dnn_predicts_test, cnn_predicts_test, gb_predicts_test], axis=1)

# bagging(avg. predict)
bag_avg_train = (dnn_predicts_train + cnn_predicts_train + gb_predicts_train) / 3
bag_avg_test = (dnn_predicts_test + cnn_predicts_test + gb_predicts_test) / 3
bag_avg_train_acc = accuracy_score(y_train, np.argmax(bag_avg_train, axis=1))
bag_avg_test_acc = accuracy_score(y_test, np.argmax(bag_avg_test, axis=1))
df_bag_avg = pd.DataFrame({
    "accuracy": [bag_avg_train_acc, bag_avg_test_acc],
    "ensamble": "bagging avg.",
    "type": ["train", "test"]
})

# bagging(weighted predict)
dnn_coef, cnn_coef, gb_coef = 1, 2, 1.5
bag_wei_train = (dnn_coef*dnn_predicts_train + cnn_coef*cnn_predicts_train + gb_coef*gb_predicts_train) / (dnn_coef + cnn_coef + gb_coef)
bag_wei_test = (dnn_coef*dnn_predicts_test + cnn_coef*cnn_predicts_test + gb_coef*gb_predicts_test) / (dnn_coef + cnn_coef + gb_coef)
bag_wei_train_acc = accuracy_score(y_train, np.argmax(bag_wei_train, axis=1))
bag_wei_test_acc = accuracy_score(y_test, np.argmax(bag_wei_test, axis=1))
df_bag_wei = pd.DataFrame({
    "accuracy": [bag_wei_train_acc, bag_wei_test_acc],
    "ensamble": "bagging weighted",
    "type": ["train", "test"]
})

# stacking(logisticregression)
lr_blender = LogisticRegression()
lr_blender.fit(predicts_train, y_train)
stack_log_train_acc = accuracy_score(y_train, np.argmax(lr_blender.predict_proba(predicts_train), axis=1))
stack_log_test_acc = accuracy_score(y_test, np.argmax(lr_blender.predict_proba(predicts_test), axis=1))
df_stack_log = pd.DataFrame({
    "accuracy": [stack_log_train_acc, stack_log_test_acc],
    "ensamble": "stacking log. regression",
    "type": ["train", "test"]
})

# stacking(tree)
tr_blender = DecisionTreeClassifier()
tr_blender.fit(predicts_train, y_train)
stack_tree_train_acc = accuracy_score(y_train, np.argmax(tr_blender.predict_proba(predicts_train), axis=1))
stack_tree_test_acc = accuracy_score(y_test, np.argmax(tr_blender.predict_proba(predicts_test), axis=1))
df_stack_tree = pd.DataFrame({
    "accuracy": [stack_tree_train_acc, stack_tree_test_acc],
    "ensamble": "stacking decision tree",
    "type": ["train", "test"]
})

df = pd.concat([df_bag_avg, df_bag_wei, df_stack_log, df_stack_tree], axis=0)
print(df)
_, ax = plt.subplots(figsize=(20, 12))
sns.barplot(x="ensamble", y='accuracy', hue='type', data=df, ax=ax)
ax.set_title("accuracy / ensambles", fontsize=20)
ax.set_ylim([df['accuracy'].min() - 0.02, 1])

del predicts_test, predicts_train, df_bag_avg, df_bag_wei, df_stack_log, df_stack_tree, bag_wei_test, bag_avg_test
```
